```python
games_file=pd.read_excel(r"C:\Users\91994\Desktop\Sprint data\Capstone project data\games.xlsx")
games_file
```

```python
game_lineup_file=pd.read_excel(r"C:\Users\91994\Desktop\Sprint data\Capstone project data\game_lineups.xlsx")
game_lineup_file
```

```python
game_event_file=pd.read_excel(r"C:\Users\91994\Desktop\Sprint data\Capstone project data\game_events.xlsx")
game_event_file
```

```python
app_file=pd.read_excel(r"C:\Users\91994\Desktop\Sprint data\Capstone project data\appearances.xlsx")
app_file
```

```python
df=app_file.merge(game_event_file,on=['game_id','player_id'],how='outer')
df1=df.merge(game_lineup_file,on=['game_id','player_id'],how='outer')
df2=df1.merge(games_file,on='game_id',how='outer')
df3=pd.merge(df2,players_file,on='player_id',how='outer')
df3
```

```python
df3.isnull().sum()
```

```python
df3.drop(columns=['home_club_formation', 'away_club_formation'],inplace=True)
```

```python
df3.dtypes
```

```python
import pandas as pd
import numpy as np

# Assuming 'df' is your dataframe
# Numerical Columns
numerical_cols = df3.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    if df3[col].isnull().sum() > 0:
        df3[col].fillna(df3[col].mean(), inplace=True)

# Categorical Columns
categorical_cols = df3.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    if df3[col].isnull().sum() > 0:
        df3[col].fillna(df3[col].mode()[0], inplace=True)

# Datetime Columns
datetime_cols = df3.select_dtypes(include=['datetime64[ns]']).columns
for col in datetime_cols:
    if df3[col].isnull().sum() > 0:
        # Replace with the median date
        df3[col].fillna(df3[col].median(), inplace=True)
```

```python
df3.isnull().sum()
```

```python
# Assuming df3 is your merged dataframe after the previous steps

# Loop through numerical columns to calculate IQR and remove outliers
for col in df3.select_dtypes(include=['number']).columns:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df3[col].quantile(0.25)
    Q3 = df3[col].quantile(0.75)
    
    # Calculate the IQR
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter out rows that are outside the bounds
    df3 = df3[(df3[col] >= lower_bound) & (df3[col] <= upper_bound)]

# Display the cleaned dataframe without outliers
print(df3)
```

```python
df3.to_csv('Football_Data_Merged_Cleaned.csv', index=False)

print("File saved as 'football_data_merged_cleaned.csv'")
```
